# Under-Sampling

# build a sample dataset containing similiar distribuation of normal transaction and fraudulent transaction

```python
legit_sample=legit.sample(n=492)
```

# concatenating two Dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

# splitting the data into features & target

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
x
```

```python
y
```

# Split the data into training dat and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
x_train_accuracy_score=accuracy_score(x_train_prediction,y_train)
```

```python
x_train_accuracy_score
```

```python
x_test_prediction=model.predict(x_test)
x_test_accuracy_score=accuracy_score(x_test_prediction,y_test)
```
